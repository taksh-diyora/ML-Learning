```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('data/package.csv')
```

```python
dataset.head()
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression as LR1
from hehe import LinearRegression as LR2
```

```python
pre = LR1()
mine = LR2()
pre.fit(X_train, y_train)
mine.fit(X_train, y_train)
```

```python
print(f"On testing data: \n Accuracy of Scikit-learn algorithm: {pre.score(X_test, y_test)} \n Accuracy of My algorithm: {mine.score(X_test, y_test)}\n")
print(f"On training data: \n Accuracy of Scikit-learn algorithm: {pre.score(X_train, y_train)} \n Accuracy of My algorithm: {mine.score(X_train, y_train)}")
```

```python
sk_pred = pre.predict(x)
my_pred = mine.predict(x)
```

```python
df = pd.DataFrame({"Sklearn Predictions" : sk_pred, "My Prediction": my_pred,"Difference": my_pred - sk_pred})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x = sk_pred, y = my_pred)
plt.xlabel("Scikit Learn Linear regression model")
plt.ylabel("My Linear regression model")
plt.savefig('image/image1.png')
plt.show()
```

```python
sns.lineplot(x = [i for i in range(len(my_pred))], y = df["Difference"])
plt.xlabel("Index of the dataframe")
plt.ylabel("Difference between My Linear Regression Predictions \n vs \n Scikit learn Linear Regression Prediction")
plt.savefig('image/image2.png')
plt.show()
```
